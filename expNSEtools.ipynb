```python
import nsepy
import pandas as pd 
import numpy as np 
```

```python
from datetime import date
```

```python
nifty100list = pd.read_csv('ind_nifty100list.csv')['Symbol']
nifty100list = np.array(nifty100list)
```

```python
from darts.timeseries import TimeSeries
```

```python
number = 1
dictClose = {}
for i in range(0, 11):
    df = pd.DataFrame(nsepy.get_history(symbol = nifty100list[i], start = date(1890, 1, 1), end = date.today() ))['Close']
    df = np.array(df)
    df = TimeSeries.from_values(df, columns=None, fillna_value=None, static_covariates=None, hierarchy=None)
    print(f"{number}: completed")
    number+=1
    dictClose[nifty100list[i]] = df
```
